```python
X_train=np.zeros((len(tlf),256,256))
```

```python
# Y_train=np.empty(len(tlf), dtype='S10')
Y_train=[]
```

```python
for i in range(len(tlf)):
    im=cv2.imread(tlf[i])
    im.resize(256,256,3)
    X_train[i,:,:]=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
```

```python
for i in range(len(tlf)):
    s=""
    dir_name=os.path.dirname(tlf[i])
    flag=0
    for char in dir_name:
            if flag==1:
                s=s+char

            if char=='\\':
                flag=1

    Y_train.append(s)
    
```
